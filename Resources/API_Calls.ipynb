```python
#import dependencies
import pandas as pd
import json
import requests

# Impor the OpenWeatherMap API key
from api_keys import geoapify_key

# Import citipy to determine the cities based on latitude and longitude
from citipy import citipy
```

```python
#read csv files
Centres_df = pd.read_csv("Centres.csv")
Centres_df
```

```python
Centres_df['Lat'] = ''
Centres_df['Lon'] = ''
Centres_df.head()
```

```python
def GetCoordinates(df):    
    base_url = "https://api.geoapify.com/v1/geocode/search?"
    params = {
                "text":"",
                 "type": "city",
                 "format":"json",
                 "apiKey": geoapify_key
            }

    for index, row in df.iterrows():

        # Create search query, make request and store in json
        params["text"] = f"{row['Centres']}, Canada"
        response = requests.get(base_url, params).json()

        try:
            df.loc[index, "Lat"] = response['results'][0]['lat']
            df.loc[index, "Lon"] = response['results'][0]['lon']
        except:
            print("City not found. Skipping...")
            pass


    # Indicate that Data Loading is complete 
    print("-----------------------------")
    print("Data Retrieval Complete      ")
    print("-----------------------------")
    
    return df
```

```python
Centres_df = GetCoordinates(Centres_df)
Centres_df.head()
```
