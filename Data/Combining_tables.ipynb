```python
#import dependencies
import pandas as pd
import json
import requests
import numpy as np
```

```python
Rental_df = pd.read_csv("Rental Data.csv")
Vacancy_df = pd.read_csv("Vacancy Rate Data.csv")
Number_of_units_df = pd.read_csv("Number of Units Data.csv")
Centres_df = pd.read_csv("Centres_latlon.csv")
Zones_df = pd.read_csv("Zones_latlon.csv")
```

```python
Rental_df.keys()
```

```python
Vacancy_df.keys()
```

```python
Number_of_units_df.keys()
```

```python
Vacancy_df = Vacancy_df.rename(columns = {  'Bachelor': 'VR Bachelor', '1 Bedroom': 'VR 1 Bedroom' , '2 Bedroom' : 'VR 2 Bedroom', '3 Bedroom +': 'VR 3 Bedroom +', 'Total' : 'VR Total'})
Vacancy_df.keys()
```

```python
Vacancy_df = Vacancy_df[['Province', 'Centres', 'Zones', 'Neighbourhood', 'Dwelling Type ',
       'VR Bachelor', 'VR 1 Bedroom', 'VR 2 Bedroom', 'VR 3 Bedroom +',
       'VR Total', 'Year', 'Centre_id', 'Zone_id']]
Vacancy_df.keys()
```

```python
Rental_df = Rental_df.rename(columns = {  'Bachelor': 'AR Bachelor', '1 Bedroom': 'AR 1 Bedroom' , '2 Bedroom' : 'AR 2 Bedroom', '3 Bedroom +': 'AR 3 Bedroom +', 'Total' : 'AR Total'})
Rental_df.keys()
```

```python
Rental_df = Rental_df[['Province', 'Centres', 'Zones', 'Neighbourhood', 'Dwelling Type ',
       'AR Bachelor', 'AR 1 Bedroom', 'AR 2 Bedroom', 'AR 3 Bedroom +', 'AR Total',
       'Year', 'Centre_id', 'Zone_id']]
Rental_df.keys()
```

```python
Number_of_units_df = Number_of_units_df.rename(columns = {  'Bachelor': 'Units Bachelor', '1 Bedroom': 'Units 1 Bedroom' , '2 Bedroom' : 'Units 2 Bedroom', '3 Bedroom +': 'Units 3 Bedroom +', 'Total' : 'Units Total'})
Number_of_units_df.keys()
```

```python
Number_of_units_df = Number_of_units_df[['Province', 'Centres', 'Zones', 'Neighbourhood', 'Dwelling Type ',
       'Units Bachelor', 'Units 1 Bedroom', 'Units 2 Bedroom',
       'Units 3 Bedroom +', 'Units Total', 'Year', 'Centre_id', 'Zone_id']]
Number_of_units_df.keys()
```

```python
Combined_df = pd.merge(Rental_df,Vacancy_df, on = ['Province','Centres','Zones','Neighbourhood','Dwelling Type ','Year', 'Centre_id', 'Zone_id'],  how = 'left')
```

```python
Combined_df = pd.merge(Combined_df,Number_of_units_df, on = ['Province','Centres','Zones','Neighbourhood','Dwelling Type ','Year', 'Centre_id', 'Zone_id'],  how = 'left')
Combined_df
```

```python
Combined_df.keys()
```

```python
ID = np.arange(1, len(Combined_df) + 1)
df_id = ['ID_' + str(each) for each in ID]
Combined_df["ID"] = df_id
Combined_df
```

```python
Combined_df.keys()
```

```python
Combined_df = Combined_df[['ID','Province', 'Centres', 'Zones', 'Neighbourhood', 'Year','Centre_id', 'Zone_id',
       'AR Bachelor', 'AR 1 Bedroom', 'AR 2 Bedroom', 'AR 3 Bedroom +',
       'AR Total', 'VR Bachelor','VR 1 Bedroom', 'VR 2 Bedroom', 'VR 3 Bedroom +', 'VR Total',
       'Units Bachelor', 'Units 1 Bedroom', 'Units 2 Bedroom',
       'Units 3 Bedroom +', 'Units Total']]
```
